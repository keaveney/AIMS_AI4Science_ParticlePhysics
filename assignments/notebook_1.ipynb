## Introduction

In this notebook, we will explore some basic concepts in statistical data analysis that are crucial for particle physics research. We will generate and visualize random variables, calculate statistical measures, and discuss their interpretations and limitations.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

## Generating and Visualizing Random Variables

Now, let's generate random numbers from a normal distribution:

```python
mu, sigma = 0, 1 # mean and standard deviation
x = np.random.normal(mu, sigma, 1000)

plt.figure(figsize=(8, 4))
plt.hist(x, bins=30, density=True, alpha=0.6, color='g')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.show()
```

## Calculating and Interpreting Basic Statistical Measures

Let's calculate the mean, median, and standard deviation of our generated data:

```python
mean = np.mean(x)
median = np.median(x)
std_dev = np.std(x)

print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}") 
print(f"Standard Deviation: {std_dev:.2f}")
```

Discuss the interpretation of these measures and how they relate to the parameters of the normal distribution.

## Quantiles and the Inverse CDF

We can calculate quantiles using the inverse cumulative density function (CDF):

```python
quantiles = [0.25, 0.5, 0.75]
for q in quantiles:
    x_q = stats.norm.ppf(q, loc=mu, scale=sigma)
    print(f"{q*100}th Percentile: {x_q:.2f}")
```

Discuss the meaning of quantiles and their usefulness in understanding the distribution of data.

## Limitations of Mean and Standard Deviation

Let's generate random numbers from a Cauchy distribution, which has heavy tails:

```python
x_cauchy = stats.cauchy.rvs(loc=0, scale=1, size=1000)

mean_cauchy = np.mean(x_cauchy)
std_dev_cauchy = np.std(x_cauchy)

print(f"Cauchy Mean: {mean_cauchy:.2f}")
print(f"Cauchy Standard Deviation: {std_dev_cauchy:.2f}")
```

Discuss why the mean and standard deviation are not well-defined for heavy-tailed distributions and introduce robust alternatives like the median and median absolute deviation (MAD).

## Application: Analyzing a Simulated Particle Physics Dataset

Generate a toy dataset with signal and background events:

```python
# Simulated dataset
np.random.seed(42)
n_signal = 1000
n_background = 10000

signal_mass = np.random.normal(125, 5, n_signal)
background_mass = np.random.exponential(50, n_background)

masses = np.concatenate((signal_mass, background_mass))
labels = np.concatenate((np.ones(n_signal), np.zeros(n_background)))
```

Calculate and interpret basic statistical measures for the dataset:

```python
# Analysis
signal_mean = np.mean(signal_mass)
signal_std = np.std(signal_mass)
background_mean = np.mean(background_mass)
background_std = np.std(background_mass)

print(f"Signal Mean: {signal_mean:.2f}")
print(f"Signal Standard Deviation: {signal_std:.2f}")
print(f"Background Mean: {background_mean:.2f}")
print(f"Background Standard Deviation: {background_std:.2f}")

plt.figure(figsize=(8, 4))
plt.hist(signal_mass, bins=30, alpha=0.6, label='Signal')
plt.hist(background_mass, bins=30, alpha=0.6, label='Background')
plt.xlabel('Mass')
plt.ylabel('Counts')
plt.legend()
plt.show()
```
